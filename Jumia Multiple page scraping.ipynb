```python
from bs4 import BeautifulSoup
import requests
import time
import datetime
import pandas as pd
```

```python
base_url= 'https://www.jumia.com.ng/samsung/'

headers= {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/135.0.0.0 Safari/537.36 Edg/135.0.0.0", "Accept-Encoding":"gzip, deflate", "Accept-Language": "en-US,en;q=0.9", "Accept": "text/html,application/xhtml+xml;q=0.9,*/*;q=0.8", "DNT":"1", "Connection":"keep-alive", "Upgrade-Insecure-Requests":"1"}

all_products= []

for page in range(1,51):
    print(f"scraping page {page}...")
    url= base_url + str(page)
    response= requests.get(url, headers=headers)
    soup1= BeautifulSoup(response.content, "html.parser")

    products= soup1.find_all('article', class_= 'prd')

    for product in products:
        title_tag= product.find('h3', class_= 'name')
        price_tag= product.find('div', class_= 'prc')
        discount_tag= product.find('div', class_= 'bdg _dsct _sm')
        ratings_tag= product.find('div', class_='stars')

        title= title_tag.text.strip() if title_tag else "N/A"
        price= price_tag.text.strip()[1:] if price_tag else "N/A"
        discount= discount_tag.text.strip() if discount_tag else "No discount"
        ratings= ratings_tag.text.strip() if ratings_tag else "No ratings"

        all_products.append({
        'Title': title,
        'Price': price,
        'Discount': discount,
        'Ratings': ratings
        })

        
df= pd.DataFrame(all_products)
df.to_csv('jumia_samsung_products.csv', index= False, encoding= 'utf-8')
print("✅ Done! Data saved as 'jumia_samsung_products.csv'")
```
